```python
import pandas as pd
import json
```

```python
contents = open('../server/answers.txt', 'r').read()

decoder = json.JSONDecoder()
i = 0
answers = []
while i < len(contents):    
    a, n = decoder.raw_decode(contents[i:])
    i += n
    answers.append(a)
```

```python
df = pd.DataFrame([a for l in answers for a in l])
df
```

```python
questions = df.question.unique()
```

```python
for q in questions:
    print('QUESTION')
    print('.'*30)
    print(q)
    print('.'*30 +"\n")
    
    df2 = df[df.question == q]
    df3 = pd.DataFrame(df2.answer.tolist())
    
    for _, row in df3.iterrows():
        print('.'*5)
        print('EXPLANATION: ' + row.errorExplanation)
        print('INTERPRETATION: ' + row.messageInterpretation)
        print('.'*5 + "\n")
        
    for p in df3.safetyViolation.tolist():
        print(p + '\n')
        
    print('.'*10+'\n')
        
    for p in df3.functionFix.tolist():
        print(p + '\n')
    
```
